```python
import torch
import torch.nn as nn


class scSenNeuNet(nn.Module):
    def __init__(self, num_genes, embedding_dim, dropout=0.1):
        super(scSenNeuNet, self).__init__()
        
        self.gene_embedding = nn.Embedding(num_genes, embedding_dim)
        self.expression_embedding = nn.Sequential(
            nn.Linear(1, embedding_dim),
            nn.LayerNorm(embedding_dim),
        )

 
        self.attention = nn.Linear(embedding_dim, 1)
        self.classifier = nn.Sequential(
            nn.Linear(embedding_dim * num_genes , 512), 
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(512, 256),  
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(256, 128),  
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(128, 1),  
            nn.Sigmoid()
        )       

    def forward(self, gene_ids, expression_values):
        gene_embed = self.gene_embedding(gene_ids)
        expr_embed = self.expression_embedding(expression_values.unsqueeze(-1))
        x = gene_embed + expr_embed


        attention_weights = torch.softmax(self.attention(x), dim=1)  # Gene importance
        weighted_x = attention_weights * x 
        weighted_x = weighted_x.flatten(start_dim=1)

        logits = self.classifier(weighted_x)

        return logits, attention_weights  
embedding_dim = 128  
model = SenNet_RNA(num_genes=len(SeneSet_MSC), embedding_dim=embedding_dim)
model.cuda()

model.load_state_dict(torch.load('SenNet_RNA.torch').state_dict())

model.eval()
```

```python
GeneID=pd.read_csv("SenNet_RNA.MarkerGene",header=None)
GeneID=list(GeneID[0])
GeneID_index = {char: idx for idx, char in enumerate(GeneID)}


gene_ids = torch.stack([torch.tensor(list(GeneID_index.values()))]*100).cuda()
```

```python
v=[]
predict=[]
corr=[]
for i in range(0,Valid_Data.shape[0]//100*100,100):
    ranID=list(range(i,i+100))
    expression_values = torch.tensor(Valid_Data[ranID,], dtype=torch.float).cuda()  
    v.extend(Valid_Target[ranID])
    with torch.no_grad():
        Class,att = model(gene_ids, expression_values)
    predict.extend(list(Class[:,0].detach().cpu().numpy()))
    corr.append(np.corrcoef(list(Class[:,0].detach().cpu().numpy()), Valid_Target[ranID])[0,1])


df=pd.DataFrame.from_dict({"v":v,"predict":predict})

plt.figure(figsize=(3, 3))
sns.violinplot(x='v', y='predict', data=df)
plt.xlabel("")
plt.ylabel("SenNeuNet")
plt.title("Pearson's R "+str(np.mean(corr))+" n=19300 ")
```
