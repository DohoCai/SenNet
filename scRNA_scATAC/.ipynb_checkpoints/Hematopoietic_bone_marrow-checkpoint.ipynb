```r
library(Seurat)
library(dplyr)
library(reticulate)
library(sctransform)
library(cowplot)
library(ggplot2)
library(viridis)
library(tidyr)
library(magrittr)
library(reshape2)
library(readxl)
# library(progeny)
library(readr)
library(stringr)
```

```r
load("Aging_Hematopoietic_BoneMarrow.RData")#Please download it from GEO
```

```r
#Figure 6A
Idents(Hematopoietic_bone_marrow)="celltype"
DimPlot(Hematopoietic_bone_marrow, label = TRUE)
```

```r
#Figure 6A
Idents(Hematopoietic_bone_marrow)="Type"
DimPlot(Hematopoietic_bone_marrow, reduction = "umap", cols = c("#6495ED", "#9ACD32", "#FF6347"))
```

```r
#Figure 6B
Hematopoietic_bone_marrow@meta.data$Type <- factor(Hematopoietic_bone_marrow@meta.data$Type, levels = c("Young", "Aged", "Ezh2cKO"))
Hematopoietic_bone_marrow@meta.data$celltype <- factor(Hematopoietic_bone_marrow@meta.data$celltype, levels = c("HSC_MPP", "GMP_EarlyGn", "LateGN","MEP_EryP","LateEry","ProMo","Meg","Lym"))
cell_fraction=table(Hematopoietic_bone_marrow@meta.data[,c("Type","celltype")])/rowSums(table(Hematopoietic_bone_marrow@meta.data[,c("Type","celltype")]))
cell_fraction
cell_fraction_d=as.data.frame(cell_fraction*100)


library(ggplot2)
p=ggplot(cell_fraction_d, aes(x = celltype, y = Freq, fill = Type)) +
  geom_bar(stat = "identity", position = "dodge") +
  theme_classic() +
  labs(title = "", x = "", y = "Percentage (%)")+
 theme(
    plot.title = element_text(size = 20),
    axis.title = element_text(size = 16),
    axis.text = element_text(size = 14),
    legend.text = element_text(size = 14),       # increase legend text size
    legend.title = element_text(size = 16),      # increase legend title size
    legend.key.size = unit(1.5, "cm")   
  ) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))+
  scale_fill_manual(values = c("#6495ED", "#9ACD32", "#FF6347")) 
p
```

```r
#Figure 6C
options(repr.plot.width = 7, repr.plot.height = 3)
your_seurat_object <- AddModuleScore(
  object = Hematopoietic_bone_marrow,
  features = list(c("Mki67", "Cdk1", "Top2a", "Pcna")),   # Note: must be a list of vectors
  name = "Proliferating"
)
Idents(your_seurat_object) <-"celltype" 
VlnPlot(your_seurat_object, features = "Proliferating1", split.by = "Type", group.by = "celltype",pt.size = 0, combine = FALSE,cols=c("#6495ED", "#9ACD32", "#FF6347"))


your_seurat_object <- AddModuleScore(
  object = Hematopoietic_bone_marrow,
  features = list(c("Itga2b", "Itga2","Slamf1","Elane",'Itgam' )),   # Note: must be a list of vectors
  name = "Myeloid"
)
Idents(your_seurat_object) <-"celltype" 
VlnPlot(your_seurat_object, features = "Myeloid1", split.by = "Type", group.by = "celltype",pt.size = 0, combine = FALSE,cols=c("#6495ED", "#9ACD32", "#FF6347"))


your_seurat_object <- AddModuleScore(
  object = Hematopoietic_bone_marrow,
  features = list(c("Tcf3", "Satb1",'Rag1', 'Il7r', 'Dntt', 'Ebf1', 'Bcl11a', 'Flt3' )),   # Note: must be a list of vectors
  name = "Lymphoid"
)
Idents(your_seurat_object) <-"celltype" 
VlnPlot(your_seurat_object, features = "Lymphoid1", split.by = "Type", group.by = "celltype",pt.size = 0, combine = FALSE,cols=c("#6495ED", "#9ACD32", "#FF6347"))


your_seurat_object <- AddModuleScore(
  object = Hematopoietic_bone_marrow,
  features = list(c( 'Cdkn1a', 'Cdkn2a', 'Ccnd1', 'Gadd45b'  )),   # Note: must be a list of vectors
  name = "Senescent"
)
Idents(your_seurat_object) <-"celltype" 
VlnPlot(your_seurat_object, features = "Senescent1", split.by = "Type", group.by = "celltype",pt.size = 0, combine = FALSE,cols=c("#6495ED", "#9ACD32", "#FF6347"))


your_seurat_object <- AddModuleScore(
  object = Hematopoietic_bone_marrow,
  features = list(c( 'Cxcr4', 'Vcam1',  'Itgb4'  )),   # Note: must be a list of vectors
  name = "Cell_Adhesion_Niche"
)
Idents(your_seurat_object) <-"celltype" 
VlnPlot(your_seurat_object, features = "Cell_Adhesion_Niche1", split.by = "Type", group.by = "celltype",pt.size = 0, combine = FALSE,cols=c("#6495ED", "#9ACD32", "#FF6347"))


```
